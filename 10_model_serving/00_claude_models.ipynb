# A\ Models on Azure Databricks

## Learnings
- Need paid (Premium) workspace to use the models
- If use the OpenAI SDK, Responses API does not work. It requires the ChatCompletion API
- If use the A\ SDK, need to add `/anthropic` to the endpoint

### Set up

```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)

DBX_TOKEN = os.getenv("DBX_TOKEN")
DBX_ENDPOINT = os.getenv("DBX_ENDPOINT")
```

### Using OpenAI SDK

```python
from openai import OpenAI

client = OpenAI(
    api_key=DBX_TOKEN,
    base_url="https://adb-7405605376511143.3.azuredatabricks.net/serving-endpoints",
)
```

```python
response = client.chat.completions.create(
    model="databricks-claude-sonnet-4-6",
    messages=[
        {
            "role": "user",
            "content": "What is your name?"
        }
    ],
    max_tokens=500
)
```

```python
print(response.choices[0].message.content)
```

### Using A\ SDK

```python
from anthropic import Anthropic

client = Anthropic(
    base_url=f"{DBX_ENDPOINT}/anthropic",
    api_key="unused",
    default_headers={
        "Authorization": f"Bearer {DBX_TOKEN}",
    },
)
```
